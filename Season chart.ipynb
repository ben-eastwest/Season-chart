```python
import yfinance as yf 
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from datetime import datetime, timedelta
```

```python
def season_chart(market):
    
    start_date = '2020-01-01'

    df = yf.download(market, start = start_date)
    df['Return'] = df['Close'].pct_change()    
    df['Year'] = df.index.year
    df['DayOfYear'] = df.index.dayofyear
    average_by_day = df.groupby(['DayOfYear', 'Year'])['Return'].mean().reset_index()
    overall_average = average_by_day.groupby('DayOfYear')['Return'].mean().reset_index()
    overall_average['Cumulative Return'] = (1 + overall_average['Return']).cumprod()

    start_date = "2024-01-01"

    data_current_year = yf.download(market, start = start_date)
    data_current_year['Return'] = data_current_year['Close'].pct_change()    
    data_current_year['Year'] = data_current_year.index.year
    data_current_year['DayOfYear'] = data_current_year.index.dayofyear    
    average_by_day_current_year = data_current_year.groupby(['DayOfYear', 'Year'])['Return'].median().reset_index()
    overall_average_current_year = average_by_day_current_year.groupby('DayOfYear')['Return'].median().reset_index()
    overall_average_current_year['Cumulative Return'] = (1 + overall_average_current_year['Return']).cumprod()
        
    # Create the traces for the seasonality and current year
    trace1 = go.Scatter(x=overall_average['DayOfYear'],y=overall_average['Cumulative Return'],mode='lines',name='Seasonality',
        line=dict(color='blue'))

    trace2 = go.Scatter(x=overall_average_current_year['DayOfYear'],y=overall_average_current_year['Cumulative Return'],mode='lines',
        name='Current Year',line=dict(color='white'))

    # Create the figure and add the traces
    fig = go.Figure()

    fig.add_trace(trace1)
    fig.add_trace(trace2)

    # Update the layout
    fig.update_layout(title='Average Values by Day of the Year',xaxis_title='Day of the Year',yaxis_title='Cumulative Return',
        template='plotly_dark',height=700,legend=dict(title='Legend',orientation='h',yanchor='bottom',y=1.02,xanchor='right',x=1))

    # Show the figure
    fig.show()
```

```python
season_chart('ura')
```
